```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
#202331129 Daniel David Hamonangan Hutabarat
```

```python
nama_file_asli = 'gambar2.jpg' 
img_bgr = cv2.imread(nama_file_asli)
img_asli_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
ukuran_asli_in_memory = img_asli_rgb.nbytes
#202331129 Daniel David Hamonangan Hutabarat
```

```python
nama_file_lossy = 'gambar_lossy_q10.jpg'
kualitas_jpeg = 10

cv2.imwrite(nama_file_lossy, img_bgr, [cv2.IMWRITE_JPEG_QUALITY, kualitas_jpeg])

ukuran_file_lossy = os.path.getsize(nama_file_lossy)

img_lossy_bgr = cv2.imread(nama_file_lossy)
img_lossy_rgb = cv2.cvtColor(img_lossy_bgr, cv2.COLOR_BGR2RGB)
#202331129 Daniel David Hamonangan Hutabarat

```

```python
jumlah_level = 4

bin_size = 256 / jumlah_level
img_kuantisasi_rgb = (np.floor(img_asli_rgb.astype(float) / bin_size) * bin_size).astype(np.uint8)

ukuran_kuantisasi_in_memory = img_kuantisasi_rgb.nbytes
#202331129 Daniel David Hamonangan Hutabarat

```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].imshow(img_asli_rgb)
axes[0].set_title(f"Asli\nUkuran: {ukuran_asli_in_memory / 1024:.2f} KB (In-Memory)")
axes[0].axis('off')

axes[1].imshow(img_lossy_rgb)
axes[1].set_title(f"Lossy JPEG Q={kualitas_jpeg}\nUkuran: {ukuran_file_lossy / 1024:.2f} KB")
axes[1].axis('off')

axes[2].imshow(img_kuantisasi_rgb)
axes[2].set_title(f"Kuantisasi RGB ({jumlah_level} Level)\nUkuran: {ukuran_kuantisasi_in_memory / 1024:.2f} KB (In-Memory)")
axes[2].axis('off')

plt.tight_layout()
plt.show()

#202331129 Daniel David Hamonangan Hutabarat
```
